```python
import pandas as pd
import string
from itertools import combinations

with open('../data/words_250000_train.txt', 'r') as file:
    lines = file.readlines()
lines = [line.strip() for line in lines]

def create_masked_dataset_all_masks(words):
    """
    Creates a dataset from a list of words.
    
    For each word, generate examples by masking all possible non-empty subsets 
    of its distinct letters. Each example replaces all occurrences of the letters 
    in the chosen subset with '_'. 
    
    The 'masked_letter' for each example is defined as the most frequent letter 
    (according to a predefined frequency order) among the masked letters.
    
    Also, for each word, compute a one-hot encoded wrong_guesses vector of size 26.
    We mark letters (from the frequency order) that are not in the word as wrong 
    guesses until 6 such letters are marked.
    
    Returns:
        A list of tuples in the form (masked_word, masked_letter, original_word, wrong_guesses)
    """
    dataset = []
    frequency_order = "etaoinshrdlcumwfgypbvkjxqz"
    alphabet = string.ascii_lowercase
    
    for k, word in enumerate(words):
        word = word.strip().lower()
        if not word:
            continue
        wrong_guesses = [0] * 26 
        wrong_count = 0
        for letter in frequency_order:
            if letter not in word:
                wrong_guesses[alphabet.index(letter)] = 1
                wrong_count += 1
                if wrong_count == 6:
                    break
        distinct_letters = sorted(set(word))
        
        for r in range(1, len(distinct_letters) + 1):
            for masked_letters_tuple in combinations(distinct_letters, r):
                masked_letters = set(masked_letters_tuple)
                masked_word = "".join('_' if ch in masked_letters else ch for ch in word)
                next_guess = next((letter for letter in frequency_order if letter in masked_letters), None)
                if next_guess is None:
                    next_guess = ""
                dataset.append((masked_word, next_guess, word, wrong_guesses.copy()))
        
        if k % 1000 == 0:
            print(f"Processed {k} words...")
            
    return dataset

df = pd.DataFrame(create_masked_dataset_all_masks(lines), 
                  columns=['masked_word', 'masked_letter', 'original_word', 'wrong_guesses'])
df.to_csv('../data/big_dataset.csv', index=False)
```
